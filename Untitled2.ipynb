```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import cKDTree
from scipy.linalg import solve
import time

# =================================================================
# 1. LRBF ENGINE CLASS
# =================================================================
class LRBFMSolver:
    def __init__(self, nodes, n_neighbors=20, shape_param=0.5):
        self.nodes = nodes  # (N, 2) array of coordinates
        self.N = nodes.shape[0]
        self.n_neighbors = n_neighbors
        self.c = shape_param # Shape parameter for Multiquadric

        # Build KD-Tree for fast local neighbor search
        self.tree = cKDTree(nodes)

        # Differentiation matrices (Sparse-like structure)
        self.weights_x = []
        self.weights_y = []
        self.weights_L = [] # Laplacian
        self.neighbor_indices = []

        print(f"Initializing LRBF with {self.N} nodes and n={n_neighbors} neighbors...")
        self._compute_all_weights()

    def _mq_kernel(self, r):
        """Multiquadric RBF: phi = sqrt(r^2 + c^2)"""
        return np.sqrt(r**2 + self.c**2)

    def _compute_all_weights(self):
        """Compute weights for first and second derivatives for every node."""
        for i in range(self.N):
            # Find local neighbors
            dist, idx = self.tree.query(self.nodes[i], k=self.n_neighbors)
            self.neighbor_indices.append(idx)

            # Extract local coordinates relative to center node i
            local_nodes = self.nodes[idx]

            # Construct the local interpolation matrix A
            A = np.zeros((self.n_neighbors, self.n_neighbors))
            for row in range(self.n_neighbors):
                for col in range(self.n_neighbors):
                    r = np.linalg.norm(local_nodes[row] - local_nodes[col])
                    A[row, col] = self._mq_kernel(r)

            # Right Hand Sides for d/dx, d/dy, and Laplacian
            rhs_x = np.zeros(self.n_neighbors)
            rhs_y = np.zeros(self.n_neighbors)
            rhs_L = np.zeros(self.n_neighbors)

            for j in range(self.n_neighbors):
                dx = self.nodes[i,0] - local_nodes[j,0]
                dy = self.nodes[i,1] - local_nodes[j,1]
                r2 = dx**2 + dy**2
                phi = np.sqrt(r2 + self.c**2)

                # Analytical derivatives of MQ kernel
                rhs_x[j] = -dx / phi
                rhs_y[j] = -dy / phi
                rhs_L[j] = (dx**2 + dy**2 + 2*self.c**2) / (phi**3)

            # Solve for weights: A * w = rhs
            self.weights_x.append(solve(A, rhs_x))
            self.weights_y.append(solve(A, rhs_y))
            self.weights_L.append(solve(A, rhs_L))

    def apply_operator(self, u, op='x'):
        """Apply the LRBF derivative operator to a field u."""
        du = np.zeros(self.N)
        if op == 'x': weights = self.weights_x
        elif op == 'y': weights = self.weights_y
        else: weights = self.weights_L

        for i in range(self.N):
            idx = self.neighbor_indices[i]
            du[i] = np.dot(weights[i], u[idx])
        return du

# =================================================================
# 2. GRID GENERATION (Boundary Layer Stretched Grid)
# =================================================================
def create_swbli_nodes(nx=40, ny=30):
    """Creates a node distribution clustered near the bottom (wall)."""
    x = np.linspace(0, 2, nx)
    # Exponential stretching for the boundary layer (y-direction)
    y = (np.exp(np.linspace(0, 1, ny)) - 1) / (np.e - 1) * 0.5

    X, Y = np.meshgrid(x, y)
    nodes = np.vstack([X.ravel(), Y.ravel()]).T
    return nodes, X, Y

# =================================================================
# 3. MAIN SIMULATION (2D Non-Linear Convection-Diffusion)
# =================================================================
# This simulates the physics of a flow hitting a wall with viscosity
nodes, X, Y = create_swbli_nodes(nx=40, ny=30)
solver = LRBFMSolver(nodes, n_neighbors=15, shape_param=0.1)

# Initial Condition: A Gaussian "pulse" mimicking a high-pressure zone
u = np.exp(-((nodes[:,0]-0.5)**2 + (nodes[:,1]-0.1)**2) / 0.02)

# Simulation Parameters
dt = 0.001
steps = 200
viscosity = 0.01
velocity_x = 1.0

print("Starting time integration...")
plt.figure(figsize=(12, 5))

for step in range(steps):
    # LRBF Spatial Derivatives
    ux = solver.apply_operator(u, 'x')
    uy = solver.apply_operator(u, 'y')
    laplacian = solver.apply_operator(u, 'L')

    # Simple explicit step: u_t + v*grad(u) = nu*laplacian(u)
    # Mimics the convection-diffusion balance in Navier-Stokes
    u_new = u + dt * (-velocity_x * ux + viscosity * laplacian)

    # Boundary Condition: No-slip at y=0
    u_new[nodes[:,1] == 0] = 0

    u = u_new.copy()

    if step % 50 == 0:
        print(f"Step {step} complete.")

# =================================================================
# 4. VISUALIZATION
# =================================================================
plt.subplot(1, 2, 1)
plt.scatter(nodes[:,0], nodes[:,1], c=u, cmap='jet', s=10)
plt.title("LRBF Field: Convection-Diffusion")
plt.xlabel("x (Streamwise)")
plt.ylabel("y (Wall Normal)")
plt.colorbar(label="Flow Intensity")

plt.subplot(1, 2, 2)
# Plot a velocity profile to show boundary layer resolution
mid_x_idx = (nodes[:,0] > 0.99) & (nodes[:,0] < 1.01)
plt.plot(u[mid_x_idx], nodes[mid_x_idx, 1], 'ro-')
plt.title("Boundary Layer Profile at x=1.0")
plt.xlabel("u")
plt.ylabel("y")
plt.grid(True)

plt.tight_layout()
plt.show()
```
